```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
titan=pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
from pandas.core.algorithms import value_counts
s=round((titan['sex'].value_counts())/len(titan)*100,2)
pd.DataFrame(s)

```

```python
labels=['male','female']
sizes=titan.sex.value_counts()
fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.f%%',shadow=True, colors=['blue','pink'])

plt.show
```

```python
plt.figure
x=titan[titan.sex=='male'].plot.scatter('age','fare',color='green',label='male')
y=titan[titan.sex=='female'].plot.scatter('age','fare',color='yellow',label='female',ax=x)
```
